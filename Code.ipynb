```python
import numpy as np
import pandas as pd
```

```python
data  = pd.read_csv(r"E:\AI Engineering Course\Sample Projects\Retail Sales and Customer Insights\Sampledata.csv", encoding ='latin1')
df = pd.DataFrame(data)
df.head()
```

### **Statistical information about the data**

```python
print(df.describe())
num_sum = df.sum(numeric_only=True)

num_sum
```

### **Check for missing values and remove duplicate orders**

```python
print(df.isnull().sum())
df.drop_duplicates(inplace=True)

print(df.info())
```

### **Create new columns for order processing time, sales per unit and profit margin**

```python
df["Order processing time"] = pd.to_datetime(df["Ship Date"]) - pd.to_datetime(df["Order Date"])
df["Sales per unit"] = df["Sales"] / df["Quantity"] 
df["Profit Margin"] = (df["Profit"] / df["Sales"]) * 100

df.head(3)
```

### **Product and category insights**
#### Top 10 products by sales and by profit trends

```python
#Product and category insights
#Top 10 products by sales and by profit trends
sales_5 = df.loc[:,["Product Name", "Sales", "Profit"]].sort_values(by="Sales", ascending=False).head(5)
profit_5 = df.loc[:, ["Product Name", "Sales", "Profit"]].sort_values(by="Profit", ascending=False).head(5)

sales_5, profit_5
```

##### Loss-making products(negative profit) despite high sales

```python
loss_making = df.loc[(df["Profit"] < 0) & (df["Sales"] > 1000), ["Product Name", "Sales", "Profit"]].sort_values(by="Profit", ascending = False).head(5)

loss_making
```

### **Customer Analysis**
#### Most valuable customers

```python
df.groupby(["Customer Name", "State"]).agg({"Sales" : "sum"}).sort_values(by="Sales", ascending=False).head()
```

#### Profitability by segments

```python
df.groupby("Segment").agg({"Profit" : "sum"}).sort_values(by="Profit", ascending=False).head()
```

### **Regional insights**
#### Profit by state and region

```python
print(df.groupby(["State", "Region"]).agg({"Profit" : "sum"}).sort_values(by = "Profit", ascending = False).head())
```

#### Regions with high sales but low margins

```python
df.groupby("Region").agg({"Profit Margin" : "mean", "Sales" : "sum"}).sort_values(by = "Profit Margin").head()
```

### **Discount impact**
#### Correlation between discount and profits

```python
print(df[["Discount", "Profit"]].corr())
```

#### Sales performance at different discount levels

```python
df["Discount"].unique()
df.groupby("Discount").agg({"Sales" : "sum", "Profit" : "sum"}).sort_values(by="Sales", ascending=False)
```

#### Category vs Region insights

```python
pd.crosstab(index=df["Category"], columns=df["Region"], values=df["Sales"], aggfunc=["mean"])
```
